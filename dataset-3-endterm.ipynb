```python
ax = df['brand'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Top 10 Brands for shoes')
ax.set_xlabel('Brand')
ax.set_ylabel('Count')
```

```python
ax = df['size'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Statistics size for shoes')
ax.set_xlabel('Size')
ax.set_ylabel('Count')
```

```python
highest_priced = df.loc[df['price'].idxmax()]
lowest_priced = df.loc[df['price'].idxmin()]

print("Highest Priced shoes:")
print(highest_priced[['brand', 'price']])

print("Lowest Priced shoes:")
print(lowest_priced[['brand', 'price']])
```

```python
highest_offer_priced = df.loc[df['offer_price'].idxmax()]
lowest_offer_priced = df.loc[df['offer_price'].idxmin()]

print("Highest Offer Priced:")
print(highest_offer_priced[['brand', 'offer_price']])

print("Lowest Offer Priced:")
print(lowest_offer_priced[['brand', 'offer_price']])
```

```python
ax = df['color'].value_counts() \
    .head(10) \
    .plot(kind='bar', title='Statistics for colors')
ax.set_xlabel('Color')
ax.set_ylabel('Count')
```
