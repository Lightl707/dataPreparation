```python
df['Discount Applied'].value_counts()
```

```python
ax = df['Discount Applied'].value_counts() \
    .head(2) \
    .plot(kind='pie', title='Statistics by Discount')
```

```python
ax = df['Gender'].value_counts() \
    .head(2) \
    .plot(kind='pie', title='Statistics by Gender')
```

```python
ax = df['Season'].value_counts() \
    .head(4) \
    .plot(kind='bar', title='Statistics for Seasons')
ax.set_xlabel('Season')
ax.set_ylabel('Count')
```

```python
avg_rating_by_year = df.groupby('Category')['Review Rating'].mean()
avg_rating_by_year.plot(kind='line', figsize=(12, 8))
plt.title('Average Rating by Category')
plt.xlabel('Category')
plt.ylabel('Review Rating')
plt.show()
```

```python
ax = df['Preferred Payment Method'].value_counts() \
    .head(5) \
    .plot(kind='bar', title='Statistics for Preferred Payment Method')
ax.set_xlabel('Preferred Payment Method')
ax.set_ylabel('Count')
```
